#Calculator Menu.

```python
class Initialize():
  def __init__(self):
      
    print("\nCalculator\n")
    print("A - Sum")
    print("B - Subtract")
    print("C - Multiply")
    print("D - Divide")
    print("E - Power")
    print("F - Square Root")
    print("G - Average")
    print("H - Maximum")
    print("I - Minimum")
    print("J - Exit")

  
  def chose_option(self):
    return input("\nChose an option: ").upper()
```

# Math Operations.

```python
class Operations():
  def __init__(self, entrance = [], result = 0):
    self.entrance = entrance
    self.result = result
  

  def to_add(self):   
    for x in self.entrance:
      self.result += float(x)

    return print(f"Result: {round(self.result, 2)}")


  def subtract(self):
    self.result += int(self.entrance[0])

    for x in self.entrance[1:]:
      self.result -= float(x)

    return print(f"Result: {round(self.result, 2)}")

  
  def multiply(self):
    self.result = 1

    for x in self.entrance:
      self.result *= float(x)

    return print(f"Result: {round(self.result, 2)}")


  def to_divide(self):
    self.result = float(self.entrance[0])

    for x in self.entrance[1:]:
      if int(x) == 0:
        x = 1

      else:
        pass

      self.result /= float(x)

    return print(f"Result: {round(self.result, 2)}")


  def to_power(self):
    self.result += float(self.entrance[0])

    for x in self.entrance[1:]:
      self.result = pow(self.result, float(x))

    return print(f"Result: {round(self.result, 2)}")


  def to_square_root(self):
    self.result += float(self.entrance[0])

    for x in self.entrance[1:]:
      self.result = pow(self.result, 1/float(x))

    return print(f"Result: {round(self.result, 2)}")


  def to_average(self):
    for x in self.entrance:
      self.result += float(x)
   
    return print(f"Result: {round (self.result / len(self.entrance), 2)}")

  
  def get_maximum(self):
    _list = []
    _list = list(map(lambda x: float(x), self.entrance))

    return print(f"Result: {max(_list)}")


  def get_minimum(self):
    return print(f"Result: {min(self.entrance)}")
```

#Utils.

```python
class Utils():
  def __init__(self):
    pass

  def wrong_massage(self):
    return print("\nWrong Option!\nChose Again.")

  def choice_math(self):
    return input("\nEnter your numbers separated by spacing:\n").split()

  def error(self):
    return print(f"\nOnly numbers are allowed {error.__class__}.\nTry again!")

  def stoped(self):
    return print('\nExecution of the program interrupted by the user.')
```

#Main.

```python
while (True):
  
  util = Utils()

  try:
    validation = ['A','B','C','D','E','F','G','H','I','J']

    panel = Initialize()

    choice = panel.chose_option()
  
    output.clear()

    if choice == 'J':
      break

    elif choice not in validation:
      util.wrong_massage()

    else:
      try:
        entrance = util.choice_math()
        math = Operations(entrance)

        if choice == 'A':
          math.to_add()

        elif choice == 'B':
          math.subtract()

        elif choice == 'C':
          math.multiply()

        elif choice == 'D':
          math.to_divide()

        elif choice == 'E':
          math.to_power()

        elif choice == 'F':
          math.to_square_root()

        elif choice == 'G':
          math.to_average()

        elif choice == 'H':
          math.get_maximum()

        elif choice == 'I':
          math.get_minimum()

        else:
          pass

      except Exception as error:
        util.error()

      except KeyboardInterrupt:
        util.stoped()
        break

  except KeyboardInterrupt:
    util.stoped()
    break
```
